# MATPLOTLIB Tutorial

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
"""Use NumPy to generate random data
"""
import numpy as np

x=np.linspace(0,10,25)
y = x * x + 2
print(x)
print(y)
```

```python
""" Command to draw
"""
pylab.plot(x,y, 'r')
```

```python
"""drawing sub graphs
"""
pylab.subplot(1,2,1) #The contents of the brackets represents (rows, columns, indexes)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x,'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()
#fig=plt.figure(figsize =(12,8), dpi =300)
axes = fig.add_axes([0.5,0.1,0.8,0.8]) #Control the left, right, width anf height of the canvas (from 0 to 1)
axes.plot(x,y, 'r')
```

```python
# Drwaing sub graphs
fig, axes = plt.subplots(nrows =1, ncols =2) # Submap is of 1 row and 2 colomuns
for ax in axes :
    ax.plot(x,y, 'r')
```

```python
# Drawing a picture / graph inside another graph
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # Big axes
axes2 = fig.add_axes([0.2,0.6,0.3,0.2]) # Small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig, axes = plt.subplots(dpi = 150) # 150 DPI
axes.plot(x, x**2, color='red', alpha =.8)
axes.plot(x, x+2, color ='#1155dd', alpha =.8)
axes.plot(x, x+3, color ='green', alpha =.8)
```
